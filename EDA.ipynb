```python
def get_income(x):
    if x == 'UNDER $1 000':
        return 1
    elif x == '$1 000 TO 2 999':
        return 1000
    elif x == '$3 000 TO 3 999':
        return 3000
    elif x == '$4 000 TO 4 999':
        return 4000
    elif x == '$5 000 TO 5 999':
        return 5000
    elif x == '$6 000 TO 6 999':
        return 6000
    elif x == '$7 000 TO 7 999':
        return 7000
    elif x == '$8 000 TO 9 999':
        return 8000
    elif x == '$10000 TO 12499':
        return 10000
    elif x == '$12500 TO 14999':
        return 12500
    elif x == '$15000 TO 17499':
        return 15000
    elif x == '$17500 TO 19999':
        return 17500
    elif x == '$20000 TO 22499':
        return 20000
    elif x == '$22500 TO 24999':
        return 22500
    elif x == '$25000 TO 29999':
        return 25000
    elif x == '$30000 TO 34999':
        return 30000
    elif x == '$35000 TO 39999':
        return 35000
    elif x == '$40000 TO 49999':
        return 40000
    elif x == '$50000 TO 59999':
        return 50000
    elif x == '$60000 TO 74999':
        return 60000
    elif x == '$75000 TO $89999':
        return 75000
    elif x == '$90000 TO $109999' or x == '$90000 - $109999':
        return 90000
    elif x == '$110000 TO $129999' or x == '$110000 OR OVER':
        return 110000
    elif x == '$130000 TO $149999':
        return 130000
    elif x == '$150000 TO $169999' or x == '$150000 OR OVER':
        return 150000   
    else:
        return 170000       
```

```python
df["INCOME"] = df["INCOME"].apply(lambda x: get_income(x))
```

# POLVIEWS

```python
df_dummies_polviews = df.drop(columns = "MAJOR1")
df_dummies_polviews = df_dummies_polviews.dropna()
tmp_y = df_dummies_polviews["POLVIEWS"]
df_dummies_polviews = df_dummies_polviews.drop(columns = "POLVIEWS")
```

```python
df_dummies_polviews = pd.get_dummies(df_dummies_polviews)
```

```python
df_dummies_polviews.shape
```

```python
tmp_y.shape
```

```python
df_dummies_polviews.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_dummies_polviews, tmp_y, test_size=0.2, random_state=42)
```

#### SVM

```python
model_polviews = svm.SVC(random_state=42)
model_polviews.fit(X_train, y_train)
```

```python
model_polviews.score(X_test, y_test)
```

#### Logistic regression

```python
model_polviews = LogisticRegression(random_state=42)
model_polviews.fit(X_train, y_train)
```

```python
model_polviews.score(X_test, y_test)
```

#### Decision tree

```python
model_polviews = tree.DecisionTreeClassifier(random_state=42, max_depth=5)
model_polviews.fit(X_train, y_train)
```

```python
model_polviews.score(X_test, y_test)
```

#### Random forest

```python
model_polviews = RandomForestClassifier(max_depth=6, random_state=42)
model_polviews.fit(X_train, y_train)
```

```python
model_polviews.score(X_test, y_test)
```

#### PREDICTING POLVIEWS

```python
df_final = pd.get_dummies(df.drop(columns=["POLVIEWS", "MAJOR1"]))
```

```python
df_final["POLVIEWS_pred"] = model_polviews.predict(df_final)
df_final["POLVIEWS"] = df["POLVIEWS"]
df_final["MAJOR"] = df["MAJOR1"]
```

```python
df_final["POLVIEWS"].fillna("U", inplace=True)
```

```python
def pred_polviews(x, y):
    if x == "U":
        return y
    else:
        return x
```

```python
df_final["POLVIEWS"] = df_final.apply(lambda row : pred_polviews(row['POLVIEWS'], row['POLVIEWS_pred']), axis = 1)
```
