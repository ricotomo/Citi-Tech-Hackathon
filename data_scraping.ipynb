## Combine RINCOM98, RINCOM06, RINCOM16 and RINCOME to INCOME

```python
df['RINCOM16'][0]
```

```python
df['RINCOM98'].fillna("nan",inplace=True)
df['RINCOM06'].fillna("nan",inplace=True)
df['RINCOM16'].fillna("nan",inplace=True)
```

```python
def get_income(row):
    if row['RINCOM16']!='nan':
        return row['RINCOM16']
    elif row['RINCOM06']!='nan':
        return row['RINCOM06']
    elif row['RINCOM98']!='nan':
        return row['RINCOM98']
    else:
        return row['RINCOME']
```

```python
df['INCOME'] = df.apply(lambda row: get_income(row), axis=1)
```

```python
df['INCOME'].isnull().sum()
```

```python
df.head(60)
```

```python
#take out the columns we do not need
df.drop(['RINCOME', 'RINCOM98','RINCOM06','RINCOM16','ISCO681','ISCO88'], axis=1,inplace=True)
```

```python
#drop missing values for INCOME column
df = df[df['INCOME'].notnull()]
```

```python
#check missing values
for i in df.columns:
    print(i, df[i].isnull().sum())
```

## Select Age 21-45, Education college and above

```python
#check the values of columns that should be numeric
df.SIBS.unique()
df.EDUC.unique()
df.PAEDUC.unique()
df.MAEDUC.unique()
```

```python
# Age has 89 OR OLDER that needs to be recoded
df.AGE.unique()
```

```python
#recode 89 or older into 89
df['AGE'].loc[df['AGE'] =='89 OR OLDER'] ='89'
```

```python
df[["SIBS", "AGE", "EDUC","PAEDUC","MAEDUC"]]=df[["SIBS", "AGE", "EDUC","PAEDUC","MAEDUC"]].apply(pd.to_numeric)
```
